## Building the KnowledgeGraph
Import all the nessecory functions from GraphRag to build the graph

````python
from Project-Explainer.graph_rag.graph_builder.tools import load_directory, initialize_llm
````

````python
from Project-Explainer.graph_rag.graph_builder.knowledgeGraph import build_graph, save_index
````

### Initialization
Initialize llm with default settings that is llama3 with *chunk size=512*

````python
initialize_llm()
````

### Data Loading
Provide the path to the directory containing the documents that is keras-nlp.md which loads the file,chunks it and create a llama-index *Document* object


````python
documents = load_directory("/content/drive/MyDrive/templates")
````

### GraphCreation
now we build KnowledgeGraph index by passing the douments and setting *max_triplets_per_chunk=5* and leaving other settings to default(*embeddings=microsoft/codebert-base*).  

Index creation takes time!

````python
index = build_graph(documents, max_triplets_per_chunk=5)
````

### Saving Index
we save the index built by serealizing with pickle so that we can query on this graph later and we don't have to build KnowledgeGraphs repeatedly

````python
save_index(index)
````

### Visulaisation
build_graph() method automatically creates and Graph_visualization.html file that gives a visualisation of the graph

````python
from IPython.display import display, HTML

# Load and display the HTML file
with open("/content/Graph_visualization.html", "r") as file:
    html_content = file.read()

display(HTML(html_content))
````

````python
query_engine = index.as_query_engine(
    include_text=True,
    response_mode="tree_summarize",
    embedding_mode="hybrid",
    similarity_top_k=5,
)
````

````python
response = query_engine.query(
    "application of kerasnlp",
)
````

````python
from IPython.display import Markdown, display

display(Markdown(f"{response}"))
````

KerasNLP is a Python library that provides a simple and easy-to-use interface for applying Natural Language Processing (NLP) techniques to text data. It is built on top of the popular deep learning library Keras, which is itself an API implemented in Python, based on the TensorFlow or Theano frameworks.

The application of KerasNLP can be seen in various areas such as:

1. Sentiment Analysis: KerasNLP can be used to analyze text data and determine whether it's positive, negative, or neutral.
2. Text Classification: It can classify text into different categories like spam/not-spam emails, or news articles into different topics.
Language Modeling: KerasNLP can predict the next word in a sequence of words based on the context.
3. Named Entity Recognition (NER): It can identify named entities such as people, places, and organizations from unstructured text data.
Overall, the application of KerasNLP is in areas where there's a need to analyze or process large amounts of text data using deep learning techniques.
